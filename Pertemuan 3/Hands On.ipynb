### Data Preparation

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("worldometer_coronavirus_daily_data.csv")
df.head()
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
df.head(1000)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d', errors='ignore')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
df
```

```python
# tabel total cases & total per country

covid_percountry = df[['country', 'daily_new_cases','daily_new_deaths']].copy()
covid_percountry = covid_percountry.groupby('country', as_index=False).sum()
covid_percountry.head()
```

```python
# tabel total case & total death per month
covid_permonth = df[['date', 'daily_new_cases','daily_new_deaths']].copy()
covid_permonth = covid_permonth.groupby(pd.Grouper(key='date', axis=0, freq='M')).sum()
covid_permonth.head()
```

```python
# tabel total cases & total per country

covid_percountry_2 = df[['country', 'cumulative_total_cases','cumulative_total_deaths','active_cases']].copy()
covid_percountry_2 = covid_percountry_2.groupby('country', as_index=False)['country', 'cumulative_total_cases','cumulative_total_deaths','active_cases'].last()
```

```python
covid_percountry_2.head()
```

```python
covid_percountry_2['Total Recovered'] = covid_percountry_2['cumulative_total_cases'] - covid_percountry_2['cumulative_total_deaths']-covid_percountry_2['active_cases']
covid_percountry_2.head()
```
